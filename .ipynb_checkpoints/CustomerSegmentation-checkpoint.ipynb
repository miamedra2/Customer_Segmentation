```python
def getelbowplot(df_in):

    sse = []
    K = range(1,15)
    for k in K:
        km = KMeans(n_clusters=k)
        km = km.fit(df_in)
        sse.append(km.inertia_)
        
    plt.plot(K, sse, 'bx-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Distortion')
    plt.title('Elbow Plot')
```

```python
droplist = []
for elem in df1:
    for cnt in range(len(df1[elem].unique())):
        if np.isnan(df1[elem].unique()[cnt]) == True and len(df1[elem].unique()) <=3 or len(df1[elem].unique()) <=1:
            droplist.append(elem)
        
df1 = df1.drop(droplist, axis = 1)
```
