```python
customer_df = customer_df.drop([elem for elem in df if elem[:8] != 'Customer'], axis=1)
products_df = products_df.drop([elem for elem in df if elem[:7] != 'Product'], axis=1)
```

```python
orders_df = orders_df.drop([elem for elem in df if elem[:6] != 'Orders'], axis=1)
orderitems_df = orderitems_df.drop([elem for elem in df if elem[:11] != 'Order_Items'], axis=1)
```

```python
def ParseDataFrameNames(df_in):
    try:
        newelem = [elem.split('.')[1] for elem in df_in]
        oldelem = [elem for elem in df_in]
        for i in range(len(newelem)):
            df_in[newelem[i]] = df_in[oldelem[i]]
    
        df_in.drop(oldelem, axis=1, inplace = True)
        return df_in
    except:
        return df_in
```

```python
customer_df = ParseDataFrameNames(customer_df)
products_df = ParseDataFrameNames(products_df)
orders_df = ParseDataFrameNames(orders_df)
orderitems_df = ParseDataFrameNames(orderitems_df)
```

```python
def getelbowplot(df_in):

    Sum_of_squared_distances = []
    K = range(1,15)
    for k in K:
        km = KMeans(n_clusters=k)
        km = km.fit(df_in)
        Sum_of_squared_distances.append(km.inertia_)
        
    plt.plot(K, Sum_of_squared_distances, 'bx-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Distortion')
    plt.title('Elbow Plot')
```
